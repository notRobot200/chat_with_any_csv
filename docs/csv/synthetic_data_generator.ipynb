```python
import pandas as pd
import random
from faker import Faker

def generate_synthetic_data(n=1000):
    fake = Faker("id_ID")  
    Faker.seed(42)
    random.seed(42)
    
    data = []
    
    kategori_produk = [
        "Kesehatan", 
        "Peralatan Rumah", 
        "Elektronik", 
        "Fashion", 
        "Makanan"
        ]
    metode_pembayaran = [
        "E-Wallet", 
        "Transfer Bank", 
        "Kartu Kredit", 
        "Kartu Debit", 
        "Virtual Account", 
        "QRIS", 
        "Cash on Delivery (COD)", 
        "Kredivo",
        "Dana" 
        ]

    alamat_dict = {}
    harga_produk_dict = {}
    
    for i in range(1, n+1):
        id_transaksi = f"T{i:05d}"
        id_pengguna = f"U{random.randint(1000, 9999)}"
        id_produk = f"P{random.randint(100, 999)}"
        waktu_transaksi = fake.date_between(start_date="-1y", end_date="today").strftime("%Y-%m-%d")
        jumlah_transaksi = random.randint(1, 10)
        metode_pembayaran_terpilih = random.choice(metode_pembayaran)
        
        if id_pengguna in alamat_dict:
            alamat_pengiriman = alamat_dict[id_pengguna]
        else:
            alamat_pengiriman = fake.address().replace("\n", ", ")
            alamat_dict[id_pengguna] = alamat_pengiriman
        
        kategori_produk_terpilih = random.choice(kategori_produk)
        penilaian_produk = random.randint(1, 5)
        
        if id_produk in harga_produk_dict:
            harga_satuan = harga_produk_dict[id_produk]
        else:
            harga_satuan = random.randint(50000, 5000000)
            harga_produk_dict[id_produk] = harga_satuan
        
        total_harga = harga_satuan * jumlah_transaksi
        
        data.append([
            id_transaksi, id_pengguna, id_produk, waktu_transaksi, jumlah_transaksi,
            metode_pembayaran_terpilih, alamat_pengiriman, kategori_produk_terpilih,
            penilaian_produk, total_harga
        ])
    
    df = pd.DataFrame(data, columns=[
        "ID_Transaksi", "ID_Pengguna", "ID_Produk", 
        "Waktu_Transaksi", "Jumlah_Transaksi",
        "Metode_Pembayaran", "Alamat_Pengiriman", 
        "Kategori_Produk", "Penilaian_Produk", "Total_Harga"
    ])
    
    return df
```

```python
df_sintetis = generate_synthetic_data(n=1000)
df_sintetis.to_csv("synthetic_transaction_data.csv", index=False)
print(df_sintetis.head())
```
