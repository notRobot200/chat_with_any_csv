```python
import torch
import clip
import numpy as np

clip_model, clip_preprocess = clip.load("ViT-B/32", device="cuda" if torch.cuda.is_available() else "cpu")

def encode_images(image_paths):
    """Buat image embedding"""
    image_embeddings = []
    for img_path in image_paths:
        image = Image.open(img_path)
        image_input = clip_preprocess(image).unsqueeze(0).to("cuda" if torch.cuda.is_available() else "cpu")
        with torch.no_grad():
            image_features = clip_model.encode_image(image_input)
        image_embeddings.append(image_features.cpu().numpy())
    return np.vstack(image_embeddings)
```

```python
import fitz  
import numpy as np
import os
from PIL import Image
import io

def extract_text_images(pdf_path, output_dir="images"):
    """Ekstrak teks dan gambar dari PDF"""
    doc = fitz.open(pdf_path)
    texts = []
    os.makedirs(output_dir, exist_ok=True)
    image_paths = []

    for page_num in range(len(doc)):
        page = doc[page_num]
        texts.append(page.get_text("text"))

        for img_index, img in enumerate(page.get_images(full=True)):
            xref = img[0]
            base_image = doc.extract_image(xref)
            img_bytes = base_image["image"]
            img_path = os.path.join(output_dir, f"page_{page_num + 1}_img_{img_index + 1}.png")
            Image.open(io.BytesIO(img_bytes)).save(img_path)
            image_paths.append(img_path)
    
    return texts, image_paths

pdf_path = "docs/pdf/buku_panduan_e-office.pdf"  
texts, image_paths = extract_text_images(pdf_path)
```

```python
import faiss
import numpy as np

def build_faiss_index(text_embeds, image_embeds):
    """Gabungkan text & image embeddings dalam FAISS"""
    
    # Ensure the embeddings are in float32 format
    text_embeds = np.array(text_embeds).astype('float32')
    image_embeds = np.array(image_embeds).astype('float32')

    # Create a FAISS index for L2 (Euclidean) distance
    index = faiss.IndexFlatL2(text_embeds.shape[1])  # Assumes both embeddings have the same dimension

    # Combine the embeddings (text + image)
    combined_embeddings = np.vstack([text_embeds, image_embeds])

    # Add combined embeddings to the index
    index.add(combined_embeddings)

    return index
```
